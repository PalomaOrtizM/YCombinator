#### Data Cleaning

```python
# Drop 'Unamed: 0' columns
regions_df.drop(columns=['Unnamed: 0'], inplace=True)
badges_df.drop(columns=['Unnamed: 0'], inplace=True)
companies_df.drop(columns=['Unnamed: 0'], inplace=True)
tags_df.drop(columns=['Unnamed: 0'], inplace=True)
industries_df.drop(columns=['Unnamed: 0'], inplace=True)

#Missing values
regions_df.fillna('', inplace=True)
schools_df.fillna('', inplace=True)
badges_df.fillna('', inplace=True)
companies_df.fillna('', inplace=True)
tags_df.fillna('', inplace=True)
industries_df.fillna('', inplace=True)
founders_df.fillna('', inplace=True)
prior_companies_df.fillna('', inplace=True)
```

```python
# Verify columns
print (regions_df.columns)
print(badges_df.columns)
print(companies_df.columns)
print(tags_df.columns)
print(industries_df.columns)
```

```python
# Update CSV
regions_df.to_csv('regions_updated.csv', index=False)
badges_df.to_csv('badges_updated.csv', index=False)
companies_df.to_csv('companies_updated.csv', index=False)
tags_df.to_csv('tags_updated.csv', index=False)
industries_df.to_csv('industries_updated.csv', index=False)
```
